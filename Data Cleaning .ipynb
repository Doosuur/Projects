## 1. Rename columns

```python
# Replace columns having space inbetween with '_'
data.columns = [i.replace(' ', '_') for i in data.columns]
```

## 2. Dealing with the Weight and Height Columns

### Height Column

```python
#Check if all the heights are in 'cm'

data['Height'][data['Height'].notnull()].str.endswith('cm').all()
```

This means some values are not in 'cm' so let's find out the others.

```python
data['Height'].unique()
```

There are some heights measured in inches (") as well. Everything will be converted to 'cm'

```python
# define a function to convert height from feet and inches to cm.
def convert_height(y):
    
    # Split the height string into foot and inches parts
    if y[-1] == '"':
        y = y.replace('\"', '')
        inch = int(y[2:])
        foot = int(y[0])
        
    # Convert foot and inches to centimeters
        total_inches = foot * 12 + inch
        cm = total_inches * 2.54
        return round(cm)
    elif y[-1] == "m":
        return int(y[:-2])
    

#Apply function to Height column
data['Height'] = data['Height'].apply(convert_height)
```

```python
data['Height'].unique()
```

### Weight Column

```python
#Check if all the weight are in 'kg'

data['Weight'][data['Weight'].notnull()].str.endswith('kg').all()
```

```python
data['Weight'].unique()
```

Some weights were measured in lbs so they'll be converted to kg

```python
# define a function to convert weight from lbs to kg.
def convert_weight(y):   
    if y[-1] == 's':
        y = y.replace('lbs', '')
        lbs = int(y[:3])
        
    # Convert lbs to kg
        kg = lbs * 0.45359237
        return round(kg)
    elif y[-1] == "g":
        return int(y[:-2])
    

#Apply function to Height column
data['Weight'] = data['Weight'].apply(convert_weight)
```

```python
data['Weight'].unique()
```

## 3. Removing the '€', 'M', and 'K' symbols and converting them to appropriate values.

```python
#Check if the '€' is is the only currency in the columns

value = data['Value']
wage = data['Wage']
release_clause = data['Release_Clause']
```

```python
value.str[0].unique()
```

```python
wage.str[0].unique()
```

```python
release_clause.str[0].unique()
```

They all have the Euro sign as the only unique sign

```python
# define a function to convert value to float
def convert_value(val):
    if val[-1] == 'M':
        return float(val[1:-1]) * 1000000
    elif val[-1] == 'K':
        return float(val[1:-1]) * 1000
    else:
        return float(val[1:])

# apply function to value column
data['Value'] = data['Value'].apply(convert_value)
data['Wage'] = data['Wage'].apply(convert_value)
data['Release_Clause'] = data['Release_Clause'].apply(convert_value)

data.head()
```

## 4. Remove the '★' sign from the W/F, SM, and IR columns.

```python
data['W/F'].unique()
```

```python
data['SM'].unique()
```

```python
data['IR'].unique()
```

```python
# define a function to remove ★
def remove_star(val):
    if val[-1] == '★':
        return (val[:-1]) 
    else:
        return (val)

# apply function to value column
data['SM'] = data['SM'].apply(remove_star)
data['IR'] = data['IR'].apply(remove_star)
data['W/F'] = data['W/F'].apply(remove_star)


data.head()
```

```python
#Convert them to integert values
data['W/F'] = data['W/F'].astype(int)
data['IR'] = data['IR'].astype(int)
data['SM'] = data['SM'].astype(int)
```

## 5. Examine the Hits column for any more anomalies.

```python
data['Hits'].unique()
```

Remember that the Hits column has missing values. We'll replace the null values with 0.

```python
# replace null values with 0
data['Hits'].fillna('0', inplace=True)
```

```python
#define a function to replace the 'K' with 1000
def hits_col(val):
    val = val.lstrip()
    if val.endswith('K'):
        val = val[:-1]
        val = float(val) * 1000
        val = int(val)
    else:
        val = int(val)
    return val

data['Hits'] = data['Hits'].apply(hits_col)
```

## 6. Clean the contract column

```python
data['Contract'].unique()
```

Looks like Some players are on contracts while others are on Loan. The column will be splitted and to have the start and end of the contract.

```python
contract = data['Contract'].copy()
```

```python
#define a function to for the start and end dates of contract column.
def contract_date(val):
  
    if '~' in val:
        dates = val.split(' ~ ')
        start_year = dates[0]
        start_year = dates[1]
        return [start_year, start_year]

    else:
        return 'NAN'   
```

```python
contract = contract.apply(contract_date)
```

```python
data['start_date'] = [i[0] if len(i[0]) == 4 else np.nan for i in contract]
```

```python
data['end_date'] = [i[1] if len(i[0]) == 4 else np.nan for i in contract]
```

```python
data['end_date'].unique()
```

```python
# Add a placeholder month and day to each year, and convert to datetime
data['start_date'] = pd.to_datetime(data['start_date'] + '-01-01', format='%Y-%m-%d', errors='coerce')
data['end_date'] = pd.to_datetime(data['end_date']+ '-01-01', format='%Y-%m-%d', errors='coerce')

#df['Year'] = pd.to_datetime(df['Year'], errors='coerce')
```

```python
data['end_date'].dtype
```

## 7. Convert the joined column to a datetime column.

```python
# convert date column to datetime format
data['Joined'] = pd.to_datetime(data['Joined'], format='%b %d, %Y')

# extract the day of the month from the datetime column
data['Day'] = data['Joined'].dt.day

# extract the month from the datetime column as a name or number
data['Month_name'] = data['Joined'].dt.month_name()
data['Month_number'] = data['Joined'].dt.month

# extract the year from the datetime column
data['Year'] = data['Joined'].dt.year

data.head()
```
